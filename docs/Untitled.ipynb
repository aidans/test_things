```python
# import pandas
import pandas as pd
# import matplotlib
import matplotlib.pyplot as plt
# import seaborn
import seaborn as sns
%matplotlib inline
```

```python
marks = pd.read_csv("ds_marks_18-19.csv")
```

```python
marks_tidy = pd.melt(
    marks,
    id_vars=['Overall'],
    var_name=['Module'],
    value_name='Mark'
)
```

```python
marks_tidy.dropna()
```

```python
marks_tidy.boxplot(by='Module')
```

```python
bplot = sns.boxplot(y='Mark', x='Module', 
                 data=marks_tidy, 
                 width=0.5,
                 linewidth=1,                    
                 fliersize=0,
                 palette="colorblind")
plt.axhline(50, color="0.5", linestyle="--", linewidth=1);
bplot=sns.stripplot(y='Mark', x='Module', 
                   data=marks_tidy, 
                   jitter=False, 
                   marker='o', 
                   alpha=0.1,
                   color='black')
bplot.axes.set_title("Data Science marks for core modules (18-19)",fontsize=16)
bplot.get_figure().savefig("boxplots.png")
```

```python
vplot = sns.violinplot(y='Mark', x='Module', 
                 data=marks_tidy, 
                 width=0.5,
                 linewidth=0.5,
                       c=0.8,
                 palette="colorblind")
plt.axhline(50, color="0.5", linestyle="--", linewidth=1);
vplot.axes.set_title("Data Science marks for core modules (18-19)",fontsize=16)
```

```python
g = sns.FacetGrid(marks_tidy, col="Module", col_wrap=3)
g.map(plt.scatter, "Overall", "Mark", edgecolor="w", alpha=0.5)

for ax in g.axes:
    ax.plot((0, 100), (0, 100), c=".8", ls="--")
ax.set(xlim=(0, 100), ylim=(0, 100))
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Data Science marks for core modules compared to module average',fontsize=16)
g.fig.savefig("scatters.png")
```
